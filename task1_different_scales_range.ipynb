## here we want to show about the different freqs in the different scale
mu control the freq, the more, the more

```python
import numpy as np
import matplotlib.pyplot as plt
mu=7
# Define the function
def f(x, mu):
    return np.exp(-x**2) * np.sin(mu * x**2)

# Generate an array of x values
x = np.linspace(-2, 2, 400)
# Calculate the y values
y = f(x, mu)
# Create the plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Plot of f(x) = exp(-x^2) * sin({mu} * x^2)')
plt.grid(True)
plt.show()
```
