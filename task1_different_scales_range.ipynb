```python
import numpy as np
import matplotlib.pyplot as plt
mu=1
# Define the function
def f(x, mu):
    return np.exp(-x**2) * np.sin(mu * x**2)

# Generate an array of x values
x = np.linspace(-1, 1, 5000)
# Calculate the y values
y = f(x, mu)
# Create the plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Plot of f(x) = exp(-x^2) * sin({mu} * x^2)')
plt.grid(True)
plt.show()
```
